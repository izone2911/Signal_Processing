```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 200, 200)
y = np.sin((1/25)*np.pi*x)


plt.figure(figsize=(16, 5))
plt.plot(x, y, label='x(n) = sin(2/50*pi*n)')
plt.title(' th x(n) = sin(2/50*pi*n)')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.grid(True)
x_tick = np.linspace(0,200,26)
plt.xticks(x_tick)
for i in range(0,201):
    plt.axvline(i, color='gray', linewidth=0.5, linestyle='-')  
for i in range(-32,33):
    plt.axhline(i/32, color='gray', linewidth=0.5, linestyle='-')
plt.axvline(0, color='black', linewidth=0.5, linestyle='-') 
plt.axhline(0, color='black', linewidth=0.5, linestyle='-')
plt.axhline(1, color='black', linewidth=0.5, linestyle='-')
plt.axhline(-1, color='black', linewidth=0.5, linestyle='-')
plt.show()
```

```python
# 64 mc lng t
# xq(n)

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 200, 200)
y = np.sin((1/25)*np.pi*x)

plt.figure(figsize=(16, 5))
plt.plot(x, y, label='xq(n)')
plt.title(' th xq(n): ng mu    Mc lng t: 64')
plt.xlabel('n')
plt.ylabel('xq(n)')
plt.grid(True)

x_tick = np.linspace(0,200,26)
plt.xticks(x_tick)

y_tick = np.linspace(-1,1,9)
y_tick_delta = [r'${0}\Delta$'.format(i) for i in range(-32,33,8)]
plt.yticks(y_tick, y_tick_delta)

num_levels = 64
quantization_levels = np.linspace(-1, 1, num_levels+1)
quantized_y = []
for i in y:
    temp = 1
    temp_val = 1
    for j in quantization_levels:
        if abs(i - j) < temp:
            temp = abs(i-j)
            temp_val = j
    quantized_y.append(temp_val) 

# print(quantization_levels)
# print(quantized_y)

plt.step(x, quantized_y, label='Tn hiu lng t ha', color='red', where='post')

for i in range(0,201):
    plt.axvline(i, color='gray', linewidth=0.5, linestyle='-')  
for i in range(-32,33):
    plt.axhline(i/32, color='gray', linewidth=0.5, linestyle='-')
plt.axvline(0, color='black', linewidth=0.5, linestyle='-') 
plt.axhline(0, color='black', linewidth=0.5, linestyle='-')
plt.axhline(1, color='black', linewidth=0.5, linestyle='-')
plt.axhline(-1, color='black', linewidth=0.5, linestyle='-')

plt.show()
```

```python
# 64 mc lng t
# e(n)

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 200, 200)
y = np.sin((1/25)*np.pi*x)

plt.figure(figsize=(16, 3))

plt.title(' th e(n)   Mc lng t: 64')
plt.xlabel('n')
plt.ylabel('e(n)')
plt.grid(True)
x_tick = np.linspace(0,200,26)
plt.xticks(x_tick)

y_tick = np.linspace(-1/64,1/64,3)
y_tick_delta = [r'$-\frac{\Delta}{2}$', 0, r'$\frac{\Delta}{2}$']
plt.yticks(y_tick, y_tick_delta)

num_levels = 64
quantization_levels = np.linspace(-1, 1, num_levels+1)
quantized_y = []
for i in y:
    temp = 1
    temp_val = 1
    for j in quantization_levels:
        if abs(i - j) < temp:
            temp = abs(i-j)
            temp_val = i-j
    quantized_y.append(temp_val) 

plt.plot(x, quantized_y, label='e(n)')

# print(quantization_levels)
# print(quantized_y)

for i in range(0,201):
    plt.axvline(i, color='gray', linewidth=0.5, linestyle='-')  
plt.axvline(0, color='black', linewidth=0.5, linestyle='-') 
plt.axhline(0, color='black', linewidth=0.5, linestyle='-')
plt.axhline(1/64, color='black', linewidth=0.5, linestyle='-')
plt.axhline(-1/64, color='black', linewidth=0.5, linestyle='-')
plt.show()
```

```python
# Tnh xp x cng sut li trung bnh Pq v SQNR vi 64 mc lng t

import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(0, 200, 200)
y = np.sin((1/25)*np.pi*x)

num_levels = 64
quantization_levels = np.linspace(-1, 1, num_levels+1)
quantized_y = []
for i in y:
    temp = 1
    temp_val = 1
    for j in quantization_levels:
        if abs(i - j) < temp:
            temp = abs(i-j)
            temp_val = i-j
    quantized_y.append(temp_val) 
sum = 0
for i in quantized_y:
    sum+=i*i
sum/=200
print('Cng sut li trung bnh Pq              :',sum)
print('Cng sut li trung bnh ca tn hiu Px :',1/2)
print('SQNR tnh ton (dB)                      :',10*math.log10(0.5/(sum)),'dB')
print('SQNR l thuyt (dB)                      :',10*math.log10(3/2*2**(2*6)),'dB')
print('===> Gi tr tnh ton v gi tr theo l thuyt xp x nhau')
```
