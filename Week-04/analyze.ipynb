```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

file_path = './Recording.wav'  
samples, sample_rate = librosa.load(file_path, sr=None)  

minrange, maxrange = min(samples),max(samples)
range = maxrange - minrange
print(minrange,maxrange,range)
```

```python
# 64 mức lượng tử
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from scipy.io.wavfile import write

file_path = './Recording.wav'  
samples, sample_rate = librosa.load(file_path, sr=None)

num_levels = 64
quantization_levels = np.linspace(minrange, maxrange, num_levels+1)
quantized_y = np.empty(0)
for i in samples:
    temp = 2
    temp_val = 1
    for j in quantization_levels:
        if abs(i - j) < temp:
            temp = abs(i-j)
            temp_val = j
    quantized_y = np.append(quantized_y,temp_val)

plt.figure(figsize=(16, 5))
librosa.display.waveshow(quantized_y, sr=sample_rate)
plt.xlabel('Thời gian')
plt.ylabel('Biên độ')
plt.show()

new_samples = (quantized_y* 32767).astype(np.int16) 
write('output64.wav', sample_rate, new_samples)
```
