```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import dlti, dimpulse

a1_values = np.linspace(-1.5, 1.5, 5)
a2_values = np.linspace(-1.0, 1.0, 5)
n = 20

fig, axes = plt.subplots(len(a1_values), len(a2_values), figsize=(12, 10))
fig.suptitle("Các Đáp Ứng Xung với Các Cặp Giá Trị (a1, a2)", fontsize=12)  

for i, a1 in enumerate(a1_values):
    for j, a2 in enumerate(a2_values):
        system = dlti([1, 0, 0], [1, a1, a2])
        
        t, x_n = dimpulse(system, n=n)
        x_n = np.squeeze(x_n)
        
        ax = axes[i, j]
        ax.stem(t, x_n)
        ax.set_title(f"a1={a1:.2f}, a2={a2:.2f}", fontsize=10)  
        ax.set_xlabel("n", fontsize=8, labelpad=-20, loc='right')  
        ax.set_ylabel("x(n)", fontsize=8)  
        ax.grid(True)
        ax.tick_params(axis='both', which='major', labelsize=8)  

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
