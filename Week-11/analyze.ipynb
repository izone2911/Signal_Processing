```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cal(t, N, f0=1):
    xt = np.zeros_like(t)
    for k in range(1, N*2, 2):
        xt += (4 / (np.pi * k)) * np.sin(2 * np.pi * k * f0 * t)
    return xt
```

```python
T = 1
t = np.linspace(0, 2*T, 10000)
```

```python
plt.figure(figsize=(16, 5))
```

```python
xt = cal(t, 1)
plt.plot(t, xt)
plt.title(f'Tín hiệu tổng hợp x(t) với N = {1}')
plt.xlabel('Thời gian (t)')
plt.ylabel('x(t)')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
xt = cal(t, 5)
plt.plot(t, xt)
plt.title(f'Tín hiệu tổng hợp x(t) với N = {5}')
plt.xlabel('Thời gian (t)')
plt.ylabel('x(t)')
plt.grid()
plt.tight_layout()
plt.show()
```
